# linear regression cost function

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# defines input
x = np.array([[1,1], [1,2], [1,3]])
y = np.array([[1], [2], [3]])
theta = np.array([[0], [1]])
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y);
plt.xlabel('x axis')
plt.ylabel('y axis');

plt.subplot(1, 2, 2)
plt.scatter(x[:,1], y, s=100, c='g', marker= 'x');
plt.xlabel('x axis')
plt.ylabel('y axis');
```

```python
# 1/2(x theta-y)T(x theta-y)
def sse(x, y, theta):
    """DocString: vector """
    prediction = x @ theta #hamun khate h
    error = prediction - y 
    return 0.5 * np.sum(error**2)
```

```python
def sse(x, y, theta):
    error = x@theta - y
    return 0.5 * (error.T * error)[0,0]
```

```python
 
theta = np.array([[0], [0.5]])
plt.plot(x, y, theta)
loss = sse(x, y, theta)
print(theta.ravel(), loss)
```
